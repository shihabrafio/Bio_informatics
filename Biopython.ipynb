# Reading Multiple Fasta Files

we are using 
# Homo sapiens phosphofructokinase, liver type (PFKL), RefSeqGene on chromosome 21, 
# Homo sapiens lncRNA oncogene in liver cancer 1 (LOLI1), long non-coding RNA, 
# Homo sapiens growth factor, augmenter of liver regeneration (GFER), mRNA

```python
from Bio import SeqIO
import pandas as pd
import os

# Specify the path to the FASTA file
multiple_fasta_file = "./data/multiplesequence.fasta"

# Read the FASTA file
sequences = SeqIO.parse(multiple_fasta_file, "fasta")

# Create an empty list to store the sequence data
sequence_data = []

# Iterate over the sequences
for sequence in sequences:
    # Access the sequence ID and sequence data
    sequence_id = sequence.id
    sequence_data.append(sequence)

    # Do something with the sequence data
    print(f"Sequence ID: {sequence_id}")
    print(f"Sequence Data: {sequence.seq}")
```

```python
len(sequence_data)
```

```python
for sequence in sequence_data:
    print(sequence.id,len(sequence.seq))
```

```python
seq_ids = []
seq_lengs = []  
for record in sequence_data:
    sequence_id = record.id
    sequence_length = len(record.seq)
    seq_ids.append((sequence_id))
    seq_lengs.append((sequence_length))
```

```python
seq_ids
```

```python
seq_lengs
```

```python
# Create a dataframe
dataframe = pd.DataFrame({'Sequence ID': seq_ids, 'Sequence Length': seq_lengs})
dataframe
```
